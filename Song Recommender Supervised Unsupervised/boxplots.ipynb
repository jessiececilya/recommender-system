```python
import numpy as np; np.random.seed(42)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
spotify_df = pd.read_csv("./spotify_files/data.csv")
spotify_df.head()
spotify_df = spotify_df.drop(spotify_df.columns[[1,3,5,7,8,10,13,14,16]], axis=1)

spotify_df.head()
normalized_df=(spotify_df-spotify_df.mean())/spotify_df.std()
```

```python
graph = sns.boxplot(x="variable",y="value",data=pd.melt(normalized_df),
           orient="v")
plt.xticks(rotation=80)
plt.tight_layout()
plt.savefig('./figures/boxplot.png')
plt.show()
```
